```python
import pickle as pkl
import numpy as np
import matplotlib.pyplot as plt
from libs.helper_func import PCA
```

```python
with open("./dataset/fer2013/all_data.pkl", 'rb') as pkl_file:
    data = pkl.load(pkl_file)

with open("./dataset/fer2013/meanfaces_eigenfaces.pkl", 'rb') as pkl_file:
    eigen_data = pkl.load(pkl_file)
```

```python
emotion_id = 2
emotion_data = data['train'][6]/255.0
mean_face_test = eigen_data[6]['mean_face'].flatten()
mean_face = eigen_data[emotion_id]['mean_face'].flatten()
eigenfaces = eigen_data[emotion_id]['eigenfaces'].reshape((-1, 48*48))
```

```python
plt.imshow(mean_face.reshape(48,48),cmap='gray')
plt.show()
```

```python
emotion_data = emotion_data.reshape((-1, 48*48))
normalized_data = emotion_data - mean_face
```

```python
recontructed = np.copy(mean_face)
for i in range(eigenfaces.shape[0]):
    weight = np.dot(normalized_data[30], eigenfaces[i])
    recontructed += weight*eigenfaces[i]
plt.imshow(recontructed.reshape(48,48),cmap='gray')
plt.show()
```
