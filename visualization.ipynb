```python
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from dataloader import FER2013_loader
```

```python
train_data = FER2013_loader(dataset_type='train')
train_loader = DataLoader(train_data, batch_size=10)
images, labels = next(iter(train_loader))
print("Images shape: ", images.shape)
print("Labels shape:", labels.shape)
```

```python
idx = 0
ax = plt.subplot(1, 1, 1)
image = images.cpu().numpy()[idx]
label = labels.cpu().numpy()[idx]
ax.imshow(image, cmap='gray')
ax.set(title=train_data.label_dict[label])
plt.show()
```

```python
ax = plt.subplot(1, 1, 1)
hist, bin_edges = train_data.get_histogram()
ax.bar(bin_edges[:-1], hist, width=0.5)
ax.set(title="Data Distribution", xticks=bin_edges[:-1],xticklabels=train_data.label_dict.values())
plt.show()
```
